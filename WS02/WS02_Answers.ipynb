### PS01

#### Construct the DataFrame

[References](https://www.geeksforgeeks.org/different-ways-to-create-pandas-dataframe/)

```python
data = {
    'X': [2, 3, 5, 7, 9],
    'Y': [4, 5, 7, 10, 15]
}

df = pd.DataFrame(data)
print(df)
```

#### Calculate the linear regression equation \( y = mx + b \) using the formula.

##### Method 1 - Using sklearn

```python
# Prepare the Data for Linear Regression
x = df[['X']]
y = df['Y']
```

```python
# Create a Linear Regression Model using sklearn
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
# Train the model
model.fit(x, y)

# Evaluate the model
r2_score = model.score(x, y)
print(f"R-squared value: {r2_score}")
```

*The R-squared value measures how well the linear regression model fits the data, ranging from 0 to 1, where 1 indicates a perfect fit.*

```python
# Get the slope (m) and intercept (b)
m = model.coef_[0]
b = model.intercept_
```

```python
# Print results
print(f"Linear Regression Equation: y = {m:.2f}x + {b:.2f}")
```

##### Method 2 - Using Formula

```python
n = df.shape[0]  # Gives number of rows
n_coln = df.shape[1]  # Gives number of columns

print(f'No. of rows: {n}\nNo. of columns: {n_coln}')
```

```python
sum_x = df['X'].sum()
print(f'Summation of X: {sum_x}')

sum_y = df['Y'].sum()
print(f'Summation of Y: {sum_y}')

sum_xy = (df['X'] * df['Y']).sum()
print(f'Summation of XY: {sum_xy}')

sum_xx = (df['X'] ** 2).sum()
print(f'Summation of X^2: {sum_xx}')

sum_yy = (df['Y'] ** 2).sum()
print(f'Summation of Y^2: {sum_yy}')

sq_sum_x = df.X.sum() ** 2
print(f'(Summation of X)^2: {sq_sum_x}')

sq_sum_y = df.Y.sum() ** 2
print(f'(Summation of Y)^2: {sq_sum_y}')
```

```python
# Slope
m = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sq_sum_x)
print(f'Slope of the given dataset is {m:.2f}')

# Intercept
c = (sum_y * sum_xx - sum_x * sum_xy) / (n * sum_xx - sq_sum_x)
print(f'Intercept of the given dataset is {c:.2f}')
```

```python
print(f"Linear Regression Equation: y = {m:.2f}x + {c:.2f}")
```

#### Compute the correlation coefficient between hours of sunshine and ice cream sales.

```python
# Compute the correlation coefficient
correlation = df['X'].corr(df['Y'])
print(f"Correlation coefficient using pandas: {correlation:.2f}")
```

```python
import numpy as np

x = np.array([2, 3, 5, 7, 9])
y = np.array([4, 5, 7, 10, 15])

# Compute the correlation coefficient matrix
correlation_matrix = np.corrcoef(x, y)
print('Correlation Matrix\n', correlation_matrix)

correlation = correlation_matrix[0, 1]
print(f"\nCorrelation coefficient using numpy: {correlation:.2f}")
```

[Reference](https://byjus.com/jee/correlation-coefficient/)

```python
from math import sqrt


r = (n * sum_xy - sum_x * sum_y) / sqrt((n * sum_xx - sq_sum_x) * (n * sum_yy - sq_sum_y))
print(f"\nCorrelation coefficient: {r:.2f}")
```

#### Find the linear regression line

```python
# Plot the data and the regression line
plt.scatter(x='X', y='Y', data=df)
plt.plot(df.X, m*df.X + c, color='red')
plt.show()
```

```python
# Plot the data and the regression line
sns.scatterplot(x='X', y='Y', data=df)
sns.lineplot(x='X', y= m*df.X + c, data=df, color='red')

plt.xlabel('Hours of Sunshine')
plt.ylabel('Ice Creams Sold')
plt.title('Linear Regression Line')
plt.show()
```

#### Determine the least squares of errors

**Normal Equations**

1. `∑y = an + b∑x`

2. `∑xy = a∑x + b∑x2`

[Reference](https://byjus.com/maths/least-square-method/)

```python
# Matrix Method - AX = b

A = np.array([[n, sum_x], [sum_x, sum_xx]])
b = np.array([sum_y, sum_xy])
```

```python
# Calculate the inverse of A
A_inv = np.linalg.inv(A)
```

```python
solution = np.dot(A_inv, b)
```

```python
b0, b1 = solution
```

```python
print(f"Linear Regression Equation: y = {b1:.2f}x + {b0:.2f}")
```

```python
y_pred = b1 * df.X + b0
sse = sum((df.Y - y_pred) ** 2)
print(f'The least squared error is {sse:.2f}')
```

```python
plt.scatter(df.X, df.Y, color='skyblue')
plt.plot(df.X, y_pred, color='red')

plt.xlabel('Hours of Sunshine')
plt.ylabel('Ice Creams Sold')
plt.title('Linear Regression Line')
plt.show()
```

#### Check the sensitivity of least squares to outliers by introducing an outlier

```python
# Introduce an outlier
outlier = {'X': 15, 'Y': 30}
df_outlier = df.copy()
df_outlier = df_outlier._append(outlier, ignore_index=True)

print(df_outlier)
```

```python
# Calculate linear regression for data with outlier
sns.lmplot(x='X', y='Y', data=df_outlier)
plt.title('Data with Outlier')
plt.xlabel('Hours of Sunshine')
plt.ylabel('Ice Creams Sold')
plt.title('Linear Regression Line')
plt.show()
```

```python
from scipy.stats import linregress


slope, intercept, r_value, p_value, std_err = linregress(df_outlier['X'], df_outlier['Y'])

print(f"Linear regression equation with outlier: y = {slope:.2f}x + {intercept:.2f}")
print(f'The Std. Error is {std_err:.2f}')
```

```python
y_pred = slope * df_outlier.X + intercept
sse = sum((df_outlier.Y - y_pred) ** 2)

print(f'The least squared error is {sse:.2f}')
```

#### Draw a box plot to detect outliers and provide an inference on it

```python
# Plot box plot for original data using plt
plt.figure(figsize=(10, 6))
# plt.boxplot(df['Y'])
plt.boxplot(df['Y'], patch_artist=True)
plt.title('Box Plot of Ice Cream Sales (Original Data)')
plt.ylabel('Ice Cream Sales')
plt.show()
```

```python
# Plot box plot for data with outlier using sns
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_outlier['Y'])
sns.stripplot(x=df_outlier['Y'], color='red', jitter=0.1, size=8)
plt.title('Box Plot of Ice Cream Sales (With Outlier)')
plt.xlabel('Ice Cream Sales')
plt.show()
```

```python
print("Inference:")
print("The box plot for the original data shows the distribution of ice cream sales without any significant outliers.")
print("The box plot with the introduced outlier shows a clear deviation, with the outlier significantly higher than the rest of the data.")
print("This demonstrates the sensitivity of the least squares method to outliers, as the presence of an outlier can greatly affect the regression results.")
```
