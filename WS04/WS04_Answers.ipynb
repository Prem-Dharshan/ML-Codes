```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set random seed for reproducibility
np.random.seed(3)

# Generate training data
x = np.random.rand(100, 1) * 10  # 100 samples of input
y = x**2 - 4 * x  # Target values based on f(x)
```

```python
# Plot the data
plt.scatter(x, y, color='blue', label='Data')
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('Generated Data')
plt.legend()
plt.show()
```
